This notebook covers essential Linux commands for file management, process monitoring, and environment maintenance.

## 1. File and Directory Management
Basic operations for navigating and manipulating files.

```python
# 进入指定目录
cd /root/workspace

# 查看当前目录下文件
ls
ls -lh    # 显示文件大小、时间、权限等详细信息

# 查看文件内容
cat results.txt # cat filename：把 filename 的内容输出到终端
head -n 10 results.txt   # 查看前10行 head results.txt
tail -n 10 results.txt   # 查看最后10行 tail results.txt

# 复制文件或文件夹
cp source_file.txt target_path/
'''
cp：复制文件/目录
source_file.txt：源文件
target_path/：目标目录（复制后文件名保持不变）
'''
cp -r source_folder target_folder   # -r 用于递归复制整个文件夹

# 移动或重命名
mv old.txt new.txt    # 重命名
mv file.txt /root/workspace/  # 移动到指定路径

# 删除（需谨慎！）
rm file_name        # 删除单个文件
# rm -rf folder_name  # 强制删除整个文件夹（不可恢复）
```

## 2. Archiving and Compression
Essential for backing up datasets and model checkpoints.

```python
# 打包并压缩文件夹
tar -zcvf dataset_final.tar.gz /root/workspace/dataset/

# 解压压缩包
tar -zxvf dataset_final.tar.gz
```

## 3. GPU and System Monitoring
Monitor hardware utilization during training.

```python
# 查看 GPU 使用情况（显存、占用率）
nvidia-smi

# 查看内存使用情况
free -h

# 查看当前运行的 Python 进程
ps aux | grep python
```

## 4. Disk Space Management
Keep track of storage and clean up caches.

```python
# 查看整体磁盘空间占用
df -h

# 查看当前文件夹的总大小
du -sh

# 清理 Hugging Face 模型缓存（释放空间）
# rm -rf ~/.cache/huggingface/
```
